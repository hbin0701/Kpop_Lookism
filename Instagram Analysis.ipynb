```python
import json
import matplotlib

f = json.load(open("final.json", "r", encoding='utf=8-sig'))
```

```python
# JIMIN => 5748 comments
# 4844 => Comments
# 2802
# 5801

# Choose from 30 posts.
# 3361, 2802, 2394, 3718

from itertools import chain
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
from collections import Counter

idx = 1

my_list = list(chain.from_iterable(f[idx][:30]))

lemmatizer = WordNetLemmatizer()
stop_words_list = stopwords.words('english')
special = [',', '.', ';', ':', '!']
def clean_text(sent):
    words = word_tokenize(sent)
    final_list = [lemmatizer.lemmatize(word) for word in words]
    return [x.lower() for x in final_list if x.lower() not in stop_words_list and len(x) >= 3 and not any(s for s in x if s in special)]
    
processed_list = list(chain.from_iterable([clean_text(x) for x in my_list]))
count = Counter(processed_list)
c = count.most_common(15)

import collections
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))
plt.xlabel("Words")
plt.ylabel("Frequency")

barlist = plt.bar([x[0] for x in c], [x[1] for x in c])


for idx in range(len(barlist)):
    barlist[idx].set_color('lightsteelblue')

    
for idx in [3, 5, 7, -2]:
    barlist[idx].set_color('red')

        
for idx in [1, 6, -5, -4, -3]:
    barlist[idx].set_color('lightsalmon')

font = {'family' : 'normal',
        'weight': 'normal',
        'size'   : 17}

matplotlib.rc('font', **font)
```
